# Transformation
Here is all the transformations needed to apply some models

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("arushchillar/disneyland-reviews")

print("Path to dataset files:", path)

path = path + "/DisneylandReviews.csv"
path
```

```python
import pandas as pd

# Load file
df = pd.read_csv(path,encoding='latin-1')

df
```

```python
df['Sentiment'] = df['Rating'].apply(lambda x: 1 if x == 5 else 0)
print(df['Sentiment'].value_counts())
```

```python
# Select important columns
df_new = df[['Review_Text','Sentiment']]
df_new
```

```python
# Write to file
df_new.to_csv("data/datasetTransformed.csv", index=False)
```
